```python
import pandas as pd
import numpy as np
```

```python
p = pd.read_csv('../data/p_polylines_data.csv')
r = pd.read_csv('../data/r_polylines_data.csv')
rp = pd.read_csv('../data/rp_polylines_data.csv')
```

```python
p.head()
```

```python
p['distance'] = np.round(p['distance']*0.000621371,1)
p['total_elevation_gain'] = np.round(p['total_elevation_gain']*3.28084,1)
```

```python
p.head()
```

```python
r.head()
```

```python
r['average_hr'] = np.round(r['average_hr'],1)
r['average_watts'] = np.round(r['average_watts'],1)
r['moving_time'] = np.round(r['moving_time']*0.0166667,1)
r['average_speed'] = np.round(r['average_speed']*2.23694,1)
```

```python
r.head()
```

```python
rp.head()
```

```python
rp['average_hr'] = np.round(rp['average_hr'],1)
rp['average_watts'] = np.round(rp['average_watts'],1)
rp['moving_time'] = np.round(rp['moving_time']*0.0166667,1)
rp['average_speed'] = np.round(rp['average_speed']*2.23694,1)
rp['distance'] = np.round(rp['distance']*0.000621371,1)
rp['total_elevation_gain'] = np.round(rp['total_elevation_gain']*3.28084,1)
```

```python
rp.head()
```

```python
p.to_csv('../data/p_polylines_data.csv', index=False)
r.to_csv('../data/r_polylines_data.csv', index=False)
rp.to_csv('../data/rp_polylines_data.csv', index=False)
```
