```python
# 3D scatter on the three features over four clusters

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['r','b','g','y']

# used the scaled data to make a better represetation
X = scaled_data
labels_true = physical_data['labels'].values

for k, col in zip(range(4), colors):
    my_members = labels_true == k
    cluster_center = km.cluster_centers_[k]
    ax.scatter(X[my_members, 0], X[my_members, 1], X[my_members, 2], c=col, marker='o',s=20) 
    ax.scatter(cluster_center[0], cluster_center[1], cluster_center[2], c=col, marker='o', s=200)
# ax.view_init(azim = 30,elev = 0)
```

```python
# checking different numbers of clusters and their intertia values

k_means_4 = KMeans(init='k-means++', n_clusters=4, n_init=10)
k_means_4.fit(scaled_data)
k_means_4_labels = k_means_4.labels_
k_means_4_cluster_centers = k_means_4.cluster_centers_

k_means_5 = KMeans(init='k-means++', n_clusters=5, n_init=10)
k_means_5.fit(scaled_data)
k_means_5_labels = k_means_5.labels_
k_means_5_cluster_centers = k_means_5.cluster_centers_

k_means_6 = KMeans(init='k-means++', n_clusters=6, n_init=10)
k_means_6.fit(scaled_data)
k_means_6_labels = k_means_6.labels_
k_means_6_cluster_centers = k_means_6.cluster_centers_

k_means_7 = KMeans(init='k-means++', n_clusters=7, n_init=10)
k_means_7.fit(scaled_data)
k_means_7_labels = k_means_7.labels_
k_means_7_cluster_centers = k_means_7.cluster_centers_

figure(figsize=(8, 5))

bar(range(4),[k_means_4.inertia_, k_means_5.inertia_, k_means_6.inertia_, k_means_7.inertia_])
xticks([0.5,1.5,2.5,3.5],('k=4','k=5','k=6','k=7'))
title("Inertia");
```

```python
# looking at the Silhouette Coefficient as a measure of close the clusters are. 
# closer to 1 the better.

k_means_4_silhouette_score = metrics.silhouette_score(scaled_data, k_means_4_labels, metric='euclidean')
k_means_5_silhouette_score = metrics.silhouette_score(scaled_data, k_means_5_labels, metric='euclidean')
k_means_6_silhouette_score = metrics.silhouette_score(scaled_data, k_means_6_labels, metric='euclidean')
k_means_7_silhouette_score = metrics.silhouette_score(scaled_data, k_means_7_labels, metric='euclidean')

figure(figsize=(8, 5))

bar(range(4),[k_means_4_silhouette_score, k_means_5_silhouette_score, k_means_6_silhouette_score, k_means_7_silhouette_score])
xticks([0.5,1.5,2.5,3.5],('k=4','k=5','k=6','k=7'))
title("Silhouette Coefficient");
```

```python
# looking at some other perspectives on Kmeans clustering models
# Calculate the squared distance between each point
distance = metrics.pairwise.euclidean_distances(scaled_data,scaled_data,squared=True)

sig = 0.5
# Turn the distance into a similarity by applying a Gaussian kernel
similarity = np.exp(-distance/sig)

figure(figsize=(5, 5))

imshow(similarity);
```

```python
# more stuff to analyze

figure(figsize=(13, 8))
# reorder according to class labels
ind_4 = np.argsort(k_means_4_labels)
subplot(1,4,1)
imshow(similarity[ind_4].T[ind_4])
title('K=4 \n' + str(k_means_4_silhouette_score))

ind_5 = np.argsort(k_means_5_labels)
subplot(1,4,2)
imshow(similarity[ind_5].T[ind_5])
title('K=5 \n' + str(k_means_5_silhouette_score))

ind_6 = np.argsort(k_means_6_labels)
subplot(1,4,3)
imshow(similarity[ind_6].T[ind_6])
title('K=6 \n' + str(k_means_6_silhouette_score))

ind_7 = np.argsort(k_means_7_labels)
subplot(1,4,4)
imshow(similarity[ind_7].T[ind_7])
title('K=7 \n' + str(k_means_7_silhouette_score));
```
